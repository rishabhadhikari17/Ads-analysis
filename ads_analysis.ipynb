```python
# importing suitable libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# reading the file in google colab
data=pd.read_excel("/content/drive/MyDrive/XYZ_Ads_Airings_Data.xlsx")
data
```

```python
# describing the dataset
data.describe()
```

```python
print(len(data.value_counts("Id")))
```

```python
# establishing correlation between different attributes of the dataset
correlation=data.corr()
correlation
```

```python
#visualizaing the data
sns.countplot(data=data,x="Brand")
plt.title("count of various brands")
plt.xticks(rotation=90)
# The brand Maruti Suzuki is occuring the most in the dataset,therefore the it isaving maximum products
# whose ads are been aired maximum time by the nwtwork
```

```python
# finding the total expenditure of the brands in ads 
sns.barplot(data=data,x="Brand",y="Spend ($)")
plt.title("Expenditure of brands in ads")
plt.xticks(rotation=90)
```

```python
"""Although Mahindra and Mahindra,TaTa motors,Toyota,Hyundai Motors India have less number of ads airing, The brands are spending
 a lot more than  Maruti Suzuki"""
```

```python
# Finding Exenditure on the basis of network type
pivot_table=data.pivot_table(index=["Network Type"],values=["Spend ($)"],aggfunc="sum")
print(pivot_table)
```

```python
X=[580528125,811934437]
plt.pie(X,labels=["broadcast","cable"])
# the expenditure of the brands on the cable network type is ,uch more than that of the broadcast
```

```python
pod_aggregate=(data.groupby("Dayparts")).agg(spendings=("Spend ($)","mean"))
print(pod_aggregate)
# The average spending on the ads airing during prime time is the most than the ads airing during the othertime
```

```python
#heat map depicting the above pod_aggregate table
sns.heatmap(pod_aggregate)
```

```python
EQ_units_sum=(data.groupby("Brand")).agg(Total_EQ_units=("EQ Units","sum"))
EQ_units_sum
```

```python
# Distribution of the pod poditions
sns.kdeplot(data=data,x="Pod Position",bw_adjust=5)
# The Occurences of Pod Positions in the range 1-10 are maximum
```

```python
# counting the number of ads month-wise
sns.displot(data["Broadcast Month"])
plt.title("Month-wise ads distribution")
```

```python
# number of ads airing during the hour of the day
sns.kdeplot(data["Hour Of Day"],fill=True,bw_adjust=3,color="g")
plt.title("No. of ads airing hour-wise")
```

```python
sns.displot(data["Duration"])
plt.title("Duration of the ads")
```
